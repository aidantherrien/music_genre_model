# ðŸŽµ AI Music Genre Classification Demo

This notebook demonstrates an AI model that classifies a song by genre.  
We will:
- Load a `.mp3` file
- Visualize its waveform and extracted audio features (MFCC, Chroma, Spectral Contrast)
- Use a trained neural network to predict the song's genre
- Visualize prediction confidence as a bar chart


```python
import librosa
import librosa.display
import matplotlib.pyplot as plt

# load song
file_path = 'demo_songs/mr_brightside_killers.mp3'
y, sr = librosa.load(file_path, duration=30) # as it stands, only pulling 30 seconds from file

# display waveform 
plt.figure(figsize=(10, 4))          # set size of image
librosa.display.waveshow(y, sr=sr)    # draw waveform of audio
plt.title("Waveform")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude") 
plt.tight_layout()                  # make it pretty
plt.show()                          # display
```

```python
import numpy as np

# extract features 
mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)
chroma = librosa.feature.chroma_stft(y=y, sr=sr)
contrast = librosa.feature.spectral_contrast(y=y, sr=sr)

# plot pseudo-spectograms to inspect features
# plot mfcc
plt.figure(figsize=(10, 4))
librosa.display.specshow(mfcc, x_axis='time', sr=sr)
plt.colorbar()
plt.title('MFCC')
plt.tight_layout()
plt.show() 

# plot chroma
plt.figure(figsize=(10, 4))
librosa.display.specshow(chroma, x_axis='time', y_axis='chroma', sr=sr)
plt.colorbar()
plt.title('Chroma')
plt.tight_layout()
plt.show()

# plot spectral contrast
plt.figure(figsize=(10, 4))
librosa.display.specshow(contrast, x_axis='time', sr=sr)
plt.colorbar()
plt.title('Spectral Contrast')
plt.tight_layout()
plt.show()
```

```python
# mix extracted features into single feature vector
features = np.hstack([
    mfcc.mean(axis=1),        # shape (13,)
    chroma.mean(axis=1),      # shape (12,)
    contrast.mean(axis=1)     # shape (7,)
])                            # result: shape (32,)
features_input = features.reshape(1, -1) # shape it for model input
```

```python
from tensorflow.keras.models import load_model

# Load the model
model = load_model("models/genre_classifier_model_v4.3.keras")

# Declare the genre mapping from the label encoder
genre_mapping = {2: 'rock', 7: 'jazz', 3: 'hip-hop', 5: 'classical', 
                 1: 'blues', 0: 'metal', 4: 'pop', 6: 'country'}

features_input = np.array(features_input)  # Put input into numpy array

# Reshape the features input to match the model's expected input shape
features_input = features_input.reshape(1, 32) 

# Make a prediction
pred = model.predict(features_input)
print("Shape of prediction:", pred.shape)

print("Raw model output:", pred)

# Get the index of the predicted genre
predicted_class_index = np.argmax(pred)

# Map the predicted label to the actual genre name
predicted_genre = genre_mapping[predicted_class_index]
print("Predicted Genre: ", predicted_genre)

# Sort the predictions and get the indices of the sorted values
sorted_indices = np.argsort(pred[0])[::-1]  # Sort in descending order
sorted_genres = [genre_mapping[i] for i in sorted_indices]
sorted_confidence = pred[0][sorted_indices]

# Plot confidence bar chart, sorted by confidence
plt.figure(figsize=(8, 4))
plt.barh(sorted_genres, sorted_confidence)  # Use the sorted genres
plt.xlabel("Confidence")
plt.title("Model Prediction Confidence")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
