```python
# 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 2: Load Dataset
# Replace the file path with the correct version of your CSV
df = pd.read_csv(r"csvs\features_v1.csv")  # e.g., "features_v4.0.csv"
```

```python
# 3: Inspect Data
print("First few rows of the dataset:")
print(df.head())

print("\nColumn names:")
print(df.columns)
```

```python
# 4: Check Class Distribution

# genre label mapping 
GENRE_LABELS = ['rock', 'jazz', 'hip-hop', 'classical', 'blues', 'metal', 'pop', 'country']

if 'genre' in df.columns:
    genre_counts = df['genre'].value_counts().sort_index()

    # Map genre index to label
    genre_labels = [GENRE_LABELS[i] for i in genre_counts.index]

    print("\nGenre distribution:")
    for label, count in zip(genre_labels, genre_counts.values):
        print(f"{label}: {count}")

    # Plot with labels
    plt.figure(figsize=(10, 6))
    sns.barplot(x=genre_labels, y=genre_counts.values, palette='viridis', hue=genre_labels, legend=False)
    plt.title("Genre Class Distribution")
    plt.xlabel("Genre")
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
else:
    print("The dataset does not contain a 'genre' column.")
```
