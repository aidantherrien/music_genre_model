```python
# 4: Check Class Distribution

from config import GENRES

if 'genre_name' in df.columns:
    from collections import Counter
    from config import GENRE_TO_INDEX

    # Count actual genres in the dataset
    genre_counter = Counter(df['genre_name'])

    # Build a list of counts for all genres in GENRES (0 if not present)
    full_genre_counts = [genre_counter.get(genre, 0) for genre in GENRES]

    # Print the full genre distribution
    print("\nFull Genre Distribution (including 0s):")
    for genre, count in zip(GENRES, full_genre_counts):
        print(f"{genre}: {count}")

    # Plot full distribution
    plt.figure(figsize=(12, 6))
    sns.barplot(x=GENRES, y=full_genre_counts, palette='viridis', hue=GENRES, legend=False)
    plt.title("Genre Class Distribution")
    plt.xlabel("Genre")
    plt.ylabel("Count")
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.show()
else:
    print("The dataset does not contain a 'genre' column.")
```
