# 🎵 AI Music Genre Classification Demo

This notebook demonstrates an AI model that classifies a song by genre.  
We will:
- Load a `.mp3` file
- Visualize its waveform and extracted audio features (MFCC, Chroma, Spectral Contrast)
- Use a trained neural network to predict the song's genre
- Visualize prediction confidence as a bar chart


```python
import librosa
import librosa.display
import matplotlib.pyplot as plt

# load song
file_path = r'C:\Users\aidan\Documents\VSCode Projects\music_genre_model\data\demo_songs\mr_brightside_killers.mp3'
y, sr = librosa.load(file_path, duration=30) # as it stands, only pulling 30 seconds from file

# display waveform 
plt.figure(figsize=(10, 4))          # set size of image
librosa.display.waveshow(y, sr=sr)    # draw waveform of audio
plt.title("Waveform")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude") 
plt.tight_layout()                  # make it pretty
plt.show()                          # display
```

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.models import load_model
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Genre list (in order corresponding to neurons in the model)
GENRES = [
    "classic_rock", "alternative_rock", "alternative", "pop_punk", "punk", "soul",
    "motown", "funk", "disco", "hip-hop", "rap", "folk", "country", "pop_country",
    "fusion", "jazz", "classical", "blues", "metal", "heavy_metal", "rock", "pop", "electronic"
]

# Mapping from genre to class index
GENRE_TO_INDEX = {genre: i for i, genre in enumerate(GENRES)}

# Mapping from class index back to genre
INDEX_TO_GENRE = {i: genre for i, genre in enumerate(GENRES)}

# Load the trained model
model = load_model(r"C:\Users\aidan\Documents\VSCode Projects\music_genre_model\models\genre_classifier_model_v8.6.h5")

def extract_features(file_path, time_steps=128):
    try:
        y, sr = librosa.load(file_path, sr=None)

        # Skip silent or very short clips
        if y is None or len(y) < sr * 3 or np.max(np.abs(y)) < 0.01:
            print(f"Skipping short or silent file: {file_path}")
            return None

        # Extracting features
        mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)
        chroma = librosa.feature.chroma_stft(y=y, sr=sr)
        contrast = librosa.feature.spectral_contrast(y=y, sr=sr)
        zero_crossings = librosa.feature.zero_crossing_rate(y=y)
        spectral_rolloff = librosa.feature.spectral_rolloff(y=y, sr=sr)
        tonnetz = librosa.feature.tonnetz(y=y, sr=sr)
        rms_energy = librosa.feature.rms(y=y)

        # Stack all features into one 2D array (time x features)
        features = np.vstack([mfcc, chroma, contrast, zero_crossings, spectral_rolloff, tonnetz, rms_energy]).T

        # Ensure the feature array matches the desired time_steps
        if features.shape[0] < time_steps:
            features = np.pad(features, ((0, time_steps - features.shape[0]), (0, 0)), mode='constant')
        else:
            features = features[:time_steps, :]

        return features
    except Exception as e:
        print(f"Error processing {file_path}: {e}")
        return None

features = extract_features(file_path)

if features is not None:
    # Flatten the features into a 2D array for normalization
    features_flattened = features.reshape(-1, features.shape[-1])

    # Step 1: Normalize the features using a new scaler (fit it on the test data)
    scaler = StandardScaler()
    features_normalized = scaler.fit_transform(features_flattened)  # Fit and transform during inference

    # Step 2: Optionally, apply PCA for dimensionality reduction
    pca = PCA(n_components=41)  # Match to the expected number of features
    features_pca = pca.fit_transform(features_normalized)  # Fit and transform for PCA

    # Reshape normalized and PCA-transformed features back to match model input
    features_pca = features_pca.reshape(1, 128, features_pca.shape[1])

    # Make a prediction
    pred = model.predict(features_pca)
    print("Shape of prediction:", pred.shape)
    print("Raw model output:", pred)

    # Get the index of the highest-confidence predicted class
    predicted_class_index = np.argmax(pred)

    # Map the predicted index to the genre name
    predicted_genre = INDEX_TO_GENRE[predicted_class_index]
    print("Predicted Genre:", predicted_genre)

    # Sort prediction confidences in descending order
    sorted_indices = np.argsort(pred[0])[::-1]
    sorted_genres = [INDEX_TO_GENRE[i] for i in sorted_indices]
    sorted_confidence = pred[0][sorted_indices]

    # Plot the confidence scores
    plt.figure(figsize=(8, 4))
    plt.barh(sorted_genres, sorted_confidence)
    plt.xlabel("Confidence")
    plt.title("Model Prediction Confidence")
    plt.gca().invert_yaxis()
    plt.tight_layout()
    plt.show()
```
