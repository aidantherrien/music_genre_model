# 🎵 AI Music Genre Classification Demo

This notebook demonstrates an AI model that classifies a song by genre.  
We will:
- Load a `.mp3` file
- Visualize its waveform and extracted audio features (MFCC, Chroma, Spectral Contrast)
- Use a trained neural network to predict the song's genre
- Visualize prediction confidence as a bar chart


```python
import librosa
import librosa.display
import matplotlib.pyplot as plt

# load song
file_path = r'C:\Users\aidan\Documents\VSCode Projects\music_genre_model\data\demo_songs\naima_john_coltrane.mp3'
y, sr = librosa.load(file_path, duration=30) # as it stands, only pulling 30 seconds from file

# display waveform 
plt.figure(figsize=(10, 4))          # set size of image
librosa.display.waveshow(y, sr=sr)    # draw waveform of audio
plt.title("Waveform")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude") 
plt.tight_layout()                  # make it pretty
plt.show()                          # display
```

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
import numpy as np


# Extract features
mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)  # Using 13 MFCC coefficients
chroma = librosa.feature.chroma_stft(y=y, sr=sr)
contrast = librosa.feature.spectral_contrast(y=y, sr=sr)

# Flatten the features into a 2D array for normalization
features = np.vstack([mfcc, chroma, contrast]).T  # Transpose to get time as rows, features as columns

# Normalize features using StandardScaler
scaler = StandardScaler()
features_normalized = scaler.fit_transform(features)

# Split the normalized features back into the original feature types
mfcc_normalized = features_normalized[:, :mfcc.shape[0]].T  # First 13 columns for MFCC
chroma_normalized = features_normalized[:, mfcc.shape[0]:mfcc.shape[0]+chroma.shape[0]].T  # Next columns for Chroma
contrast_normalized = features_normalized[:, mfcc.shape[0]+chroma.shape[0]:].T  # Last columns for Spectral Contrast

# Plot pseudo-spectrograms to inspect normalized features

# Plot normalized MFCC
plt.figure(figsize=(10, 4))
librosa.display.specshow(mfcc_normalized, x_axis='time', sr=sr)
plt.colorbar()
plt.title('Normalized MFCC')
plt.tight_layout()
plt.show()

# Plot normalized Chroma
plt.figure(figsize=(10, 4))
librosa.display.specshow(chroma_normalized, x_axis='time', y_axis='chroma', sr=sr)
plt.colorbar()
plt.title('Normalized Chroma')
plt.tight_layout()
plt.show()

# Plot normalized Spectral Contrast
plt.figure(figsize=(10, 4))
librosa.display.specshow(contrast_normalized, x_axis='time', sr=sr)
plt.colorbar()
plt.title('Normalized Spectral Contrast')
plt.tight_layout()
plt.show()
```

```python
# Mix normalized features into a single feature vector (aiming for 64 features)
features = np.hstack([
    mfcc_normalized.mean(axis=1), mfcc_normalized.std(axis=1),        # shape (26,) - Normalized MFCC: mean + std
    chroma_normalized.mean(axis=1), chroma_normalized.std(axis=1),    # shape (24,) - Normalized Chroma: mean + std
    contrast_normalized.mean(axis=1), contrast_normalized.std(axis=1)  # shape (14,) - Normalized Spectral Contrast: mean + std
])                             # Result: shape (64,)

# Reshape for model input (1, 64)
features_input = features.reshape(1, -1)

# Make it a numpy array for input
features_input = np.array(features_input)

# Print the final feature vector
print("Feature vector (64 inputs):")
print(features_input)
```

```python
from tensorflow.keras.models import load_model

# Genre list (in order corresponding to neurons in the model)
GENRES = [
    "classic_rock", "alternative_rock", "alternative", "pop_punk", "punk", "soul",
    "motown", "funk", "disco", "hip-hop", "rap", "folk", "country", "pop_country",
    "fusion", "jazz", "classical", "blues", "metal", "heavy_metal", "rock", "pop", "electronic"
]

# Mapping from genre to class index
GENRE_TO_INDEX = {genre: i for i, genre in enumerate(GENRES)}

# Mapping from class index back to genre
INDEX_TO_GENRE = {i: genre for i, genre in enumerate(GENRES)}

# Load the trained model
model = load_model(r"C:\Users\aidan\Documents\VSCode Projects\music_genre_model\models\genre_classifier_model_v7.2.keras")

# Make a prediction
pred = model.predict(features_input)
print("Shape of prediction:", pred.shape)
print("Raw model output:", pred)

# Get the index of the highest-confidence predicted class
predicted_class_index = np.argmax(pred)

# Map the predicted index to the genre name
predicted_genre = INDEX_TO_GENRE[predicted_class_index]
print("Predicted Genre:", predicted_genre)

# Sort prediction confidences in descending order
sorted_indices = np.argsort(pred[0])[::-1]
sorted_genres = [INDEX_TO_GENRE[i] for i in sorted_indices]
sorted_confidence = pred[0][sorted_indices]

# Plot the confidence scores
plt.figure(figsize=(8, 4))
plt.barh(sorted_genres, sorted_confidence)
plt.xlabel("Confidence")
plt.title("Model Prediction Confidence")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
